```python
# Adding noise to a digital image using Python
# Import the required libraries
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline 
```

```python
# Read the image from your local drive
img = Image.open('/Users/rajaraju/Downloads/sample_pics/bird.jpg').convert('L') # converting the image to grayscale
```

```python
# Display the image
plt.imshow(img,cmap='gray')

#convert the image to into numpy array for easy operations
img = np.array(img)
```

```python
type(img) # the image is now converted into n dimensional array
```

```python
# Add the noise to the image
noisy_img = img * np.random.random(img.shape)
```

```python
# Display the image after adding the noise
plt.imshow(noisy_img,cmap='gray')
plt.show()
```
